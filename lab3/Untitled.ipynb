```python
#f(x)
def f(x):
    return 1/(1+x**2)
n=5,10,15
x = np.linspace(-5, 5, 100)
plt.plot(x,f(x))
```

```python
#p(x)
def interp(n,count=30):
    results=[]
    x=np.linspace(-5,5,n+1)
    vander_x=np.vander(x,n+1,True)
    coef=np.linalg.solve(vander_x,list(map(f,x)))
    check=np.linspace(-5,5,count)
    for x in check:
        res=0
        for i,c in enumerate(coef):
            res+=c*(x**i)
        results.append(res)
    return results
plt.plot(np.linspace(-5,5,30),interp(5))
plt.plot(np.linspace(-5,5,30),interp(10))
plt.plot(np.linspace(-5,5,30),interp(15))
plt.show()
function=[]
for x in np.linspace(-5,5,30):
    function.append(f(x))
plt.plot(np.linspace(-5,5,30),function)
plt.show()
```
