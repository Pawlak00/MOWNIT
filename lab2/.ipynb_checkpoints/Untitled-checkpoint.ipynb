```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=3,floatmode='fixed',suppress=True)
def jordan_gauss_pivot(n):
    A=np.random.uniform(-1,1,(n,n))
    for i in range(n):
        max_col=i
        for j in range(n):
            if i<=j:
                if np.absolute(A[j][i])>np.absolute(A[max_col][i]):
                    max_col=j
        A[[max_col,i]]=A[[i,max_col]]
        for j in range(n):
            if i != j:
                ratio = A[j][i]/A[i][i]
                for k in range(n):
                    A[j][k] = A[j][k] - ratio * A[i][k]
    return A
```

Faktoryzacja LU

```python
n=3
A=np.random.uniform(-1,1,(n,n))
print(A)
# A=np.array([[1,1,1],[4,3,-1],[3,5,3]])
B=np.copy(A)
```

```python
def LU_decomposition():
    for i in range(n):
        for j in range(i,n):
            if i != j:
                ratio = A[j][i]/A[i][i]
                for k in range(i,n):
                    A[j][k] = A[j][k] - ratio * A[i][k]
                A[j][i]=ratio
    return A
```

```python
LU=LU_decomposition()
L=np.zeros((n,n))
L=np.tril(LU)
for i in range(n):
    L[i][i]=1
U=np.triu(LU)
C=np.matmul(L,U)
print(np.linalg.norm(np.subtract(B,C)))
```

```python
def plot_solution(edges_oriented,edges_dict,solve):
    G1=nx.DiGraph()
    for e in edges_oriented.keys():
        u,v=edges_oriented[e][0],edges_oriented[e][1]
        if solve[get_edge_number(edges_dict,(u,v))[1]]<0:
            G1.add_edge(v,u)
            G1[v][u]['weight']=-solve[get_edge_number(edges_dict,(u,v))[1]]
        else:
            G1.add_edge(u,v)
            G1[u][v]['weight']=solve[get_edge_number(edges_dict,(u,v))[1]]
    poss=nx.circular_layout(G1)
    labels={}
    for e in G1.edges():
        labels[e]=round(G1[e[0]][e[1]]['weight'],3)
    s=0.0
    for v in G1.nodes:
#         add out edges
        for u in G1[v]:
            s+=G1[v][u]['weight']
#         add in edges
        for x in G1.nodes:
            if G1[x].get(v):
                s-=G1[x].get(v).get('weight',0)
    if s<1*10**-12:
        print('OK')
    else:
        print('ERROR')
    plt.figure(3,figsize=(16,16)) 
    nx.draw_networkx_nodes(G1, poss)
    nx.draw_networkx_labels(G1, poss)
    nx.draw_networkx_edges(G1,poss,width=6,arrowsize=15,edge_cmap=plt.cm.Blues.reversed())
    nx.draw_networkx_edge_labels(G1,poss,labels)
```

```python
def read_graph_from_file(file_name):
    f=open(file_name,"r")
    G=nx.Graph()
    for line in f.readlines():
        l=line.split()
        d=''
        for d1 in l[2:]:
            d+=d1
        params=ast.literal_eval(d)
        a,b=int(l[0]),int(l[1])
        G.add_edge(a,b)
        for k in params.keys():
            G[a][b][k]=params[k]
    return G
```

```python
def gen_graph_to_file(graph_name,V,E,file_name):
    G=None
    if graph_name=='ERDOS':
        G=nx.generators.random_graphs.erdos_renyi_graph(V,0.5)
    if graph_name=='CUBIC':
        G=nx.generators.small.cubical_graph()
    if graph_name=='GRID':
        G=nx.generators.lattice.grid_graph(dim=[V,V])
        V_dict={v:i for i,v in enumerate(list(G.nodes))}
        G=nx.relabel_nodes(G,V_dict)
    if graph_name=='BRIDGE':
        G2=nx.generators.random_graphs.erdos_renyi_graph(V//2,0.9)
        G1=nx.generators.random_graphs.erdos_renyi_graph(V//2,0.9)
        V_dict={v:i+(V//2) for i,v in enumerate(list(G1.nodes))}
        G1=nx.relabel_nodes(G1,V_dict)
        G=nx.compose(G2,G1)
        G.add_edge((V//2)-1,(V//2)+1)
    if graph_name=='SMALL_WORLD':
        G=nx.generators.geometric.navigable_small_world_graph(int(np.sqrt(V)))
        V_dict={v:i for i,v in enumerate(list(G.nodes))}
        G=nx.relabel_nodes(G,V_dict)
    for e in G.edges:
        G[e[0]][e[1]]['resistance']=round(np.random.uniform(-10,10),3)
        G[e[0]][e[1]]['sem']=round(np.random.uniform(-10,10),3)
    file=open(file_name,"w")
    for line in nx.generate_edgelist(G):
        file.write(line)
        file.write('\n')
gen_graph_to_file('BRIDGE',10,0,'test1.txt')
```

```python
def get_edge_number(edges_dict,edge):
    if edge in edges_dict:
        return edge,edges_dict[edge]
    else:
        edge=(edge[1],edge[0])
        if edge in edges_dict:
            return edge,edges_dict[edge]

def solve_circuit(G):
    cycles=nx.algorithms.cycles.cycle_basis(G)
    edges_dict={e:i for i,e in enumerate(G.edges)}
    edges_oriented={}
    for cycle in cycles:
        for i in range(len(cycle)):
            u,v=cycle[i],cycle[(i+1)%len(cycle)]
            edges_oriented[(u,v)]=(u,v) if (v,u) not in edges_oriented else (v,u)
    for e in G.edges:
        u,v=e[0],e[1]
        if (u,v) not in edges_oriented and (v,u) not in edges_oriented:
            edges_oriented[(u,v)]=(u,v)
    A=np.zeros((len(cycles)+len(G.nodes),len(G.edges)))
    D=np.zeros((len(cycles)+len(G.nodes),))
    for i,cycle in enumerate(cycles):
        e_sum=0.0
        for j in range(len(cycle)):
            u,v=cycle[j],cycle[(j+1)%len(cycle)]
            if edges_oriented.get((u,v))!=(u,v):
                A[i][get_edge_number(edges_dict,(u,v))[1]]=G[u][v]['resistance']
                e_sum+=-G[u][v]['sem']
            else:
                A[i][get_edge_number(edges_dict,(u,v))[1]]=G[u][v]['resistance']
                e_sum+=G[u][v]['sem']
        D[i]=e_sum
    for i in range(len(cycles),len(cycles)+len(G.nodes)):
        node=i-len(cycles)
        if node in G:
            for e in G[node]:
                if edges_oriented.get((node,e))!=(node,e):
                    A[i][get_edge_number(edges_dict,(node,e))[1]]=1
                else:
                    A[i][get_edge_number(edges_dict,(node,e))[1]]=1
    print('cykle')
    print(cycles)
    print('kierunki krawedzi')
    print(edges_oriented)
    print('slownik krawedzi')
    print(edges_dict)
    print(A,D)
    print(np.linalg.lstsq(A,D,rcond=None)[0])
    return edges_oriented,edges_dict,np.linalg.lstsq(A,D,rcond=None)[0]
```

# Metoda potencjałów węzłowych

```python
class Component:
    def __init__(self, comp_type, high_str, low_str, value):
        self.comp_type = comp_type
        self.high_str = high_str
        self.low_str = low_str
        self.high = -1
        self.low = -1
        self.value = value
    def __repr__(self):
        return str(self.comp_type) + " " + str(self.high) + " " +str(self.low) + " " + str(self.value)
class NodeHashtable:
    def __init__(self):
        self.nodes = {}
        self.nodeCount = 0
    def addToNodes(self, node_str):
        if node_str not in self.nodes:
            self.nodes[node_str] = self.nodeCount
            self.nodeCount = self.nodeCount + 1
        return self.nodes[node_str]
    def __del__(self):
        del self.nodes
def mapNodes(components):
    hashtable = NodeHashtable()
    hashtable.addToNodes('0')
    for component in components:
        component.high = hashtable.addToNodes(component.high_str)
        component.low = hashtable.addToNodes(component.low_str)
    return components, hashtable
def calculateMatrices(components, nodeCount):
    g2Count = 2
    matrixSize = nodeCount + g2Count - 1
    A = np.zeros((matrixSize, matrixSize))
    b = np.zeros(matrixSize)
    g2Index = matrixSize - g2Count
    for component in components:
        high = component.high
        low = component.low
        value = component.value
        if component.comp_type == 'R':
            if high != 0:
                A[high - 1][high - 1] = A[high - 1][high - 1] + 1/value
            if low != 0:
                A[low - 1][low - 1] = A[low - 1][low - 1] + 1/value
            if high != 0 and low != 0:
                A[high - 1][low - 1] = A[high - 1][low - 1] - 1/value
                A[low - 1][high - 1] = A[low - 1][high - 1] - 1/value
        elif component.comp_type == 'V':
            if high != 0:
                A[high - 1][g2Index] = A[high - 1][g2Index] + 1
                A[g2Index][high - 1] = A[g2Index][high - 1] + 1
            if low != 0:
                A[low - 1][g2Index] = A[low - 1][g2Index] - 1
                A[g2Index][low - 1] = A[g2Index][low - 1] - 1
            b[g2Index] = value
            g2Index = g2Index + 1
    return A, b
def solveSystem(A, b):
    x = np.linalg.solve(A, b)
    return x
voltageCount = 0
currentCount = 0
resistorCount = 0
capacitorCount = 0
inductorCount = 0
components = parseFile(fileName)
components, hashtable = mapNodes(components)
A, b = calculateMatrices(components, hashtable.nodeCount)
x = solveSystem(A, b)
print("x:\n", x)
```
