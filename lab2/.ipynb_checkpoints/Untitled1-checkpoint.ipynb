```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=3,floatmode='fixed',suppress=True)
def jordan_gauss_pivot(n):
    A=np.random.uniform(-1,1,(n,n))
    for i in range(n):
        max_col=i
        for j in range(n):
            if i<=j:
                if np.absolute(A[j][i])>np.absolute(A[max_col][i]):
                    max_col=j
        A[[max_col,i]]=A[[i,max_col]]
        for j in range(n):
            if i != j:
                ratio = A[j][i]/A[i][i]
                for k in range(n):
                    A[j][k] = A[j][k] - ratio * A[i][k]
    return A
```

```python
from time import perf_counter
ans=[]
begin=500
end=700
step=20
for i in range(begin,end,step):
    t1=perf_counter()
    jordan_gauss_pivot(i)
    t2=perf_counter()
    ans.append((i,t2-t1))
```

```python
print(ans)
```

```python
ans1=[]
for n in range(begin,end,step):
    t1=perf_counter()
    np.linalg.solve(np.random.uniform(-1,1,(n,n)),np.random.uniform(-1,1,n))
    t2=perf_counter()
    ans1.append((n,t2-t1))
```

```python
plt.xlabel("wymiar macierzy")
plt.ylabel("czas[s]")
plt.plot([ans[i][0] for i in range(len(ans))],[ans[i][1] for i in range(len(ans))])
plt.plot([ans1[i][0] for i in range(len(ans))],[ans1[i][1] for i in range(len(ans))])
```

```python
n=3
A=np.random.uniform(-1,1,(n,n))
print(A)
# A=np.array([[1,1,1],[4,3,-1],[3,5,3]])
B=np.copy(A)
```

```python
def LU_decomposition():
    for i in range(n):
        for j in range(i,n):
            if i != j:
                ratio = A[j][i]/A[i][i]
                for k in range(i,n):
                    A[j][k] = A[j][k] - ratio * A[i][k]
                A[j][i]=ratio
    return A
```

```python
LU=LU_decomposition()
L=np.zeros((n,n))
L=np.tril(LU)
for i in range(n):
    L[i][i]=1
U=np.triu(LU)
C=np.matmul(L,U)
print(np.linalg.norm(np.subtract(B,C)))
```

zad. 3

W poniższym programie przedstawiono implementację obydwu sposobów analizy obwodu elektrycznego, korzystając zarówno z Praw Kirchoffa jak i z metody potencjałów węzłowych. W pierwszej z nich rozwiązałem nadokreślony układ równań, korzystając z funkcji lstsq z biblioteki numpy. Tą część rozwiązania można było zastąpić rozwiązywaniem "zwyczajnego", czyli oznaczonego układu równań[Matrix analysis and applied linear algebra str.75], jednak dla sformułowanego problemu przyjęty sposób rozwiązywania układu nie prowadzi do znaczących błędów.
	Aby rozwiązać problem korzystając z Praw Kirchoffa należy znaleźć wszystkie cykle bazowe grafu, do czego użytu funkcji cycle_basis z biblioteki networkx. W tym rozwiązaniu jako kierunek krawędzi przyjęto bez straty ogólności kierunek w jakim krawędź pojawiła się w cyklu. Jeśli krawędź pojawiła się w kilku cyklach przyjęto kierunek z dowolnego z nich. Weryfikacja kierunków prądu następowała po uzyskaniu rozwiązań(krawędzie z ujemnymi prądami były odwracane). Aby sprawdzić poprawność rozwiązania sumowano wagi krawędzi wchodzących i wychodzących z wierzchołka. Otrzymana suma powinna być równa 0. Jeżeli suma jest równa zero program wypisuje komunika 'OK'
    W metodzie potencjałów węzłowych kierunki prądu na krawędziach są tożsame z kolejnościa wierzchołków krawędzi na wejściu. Kierunek płynięcia prądu jest ustalany po obliczeniu potencjałów w węzłach(Przyjęto że prąd płynie od węzła o większym potencjale do węzła o mniejszym potencjale).
    Generowanie grafu do pliku z listą sąsiedztwa odbywa się w funkcji gen_graph_to_file. Argumentami jest nazwa generowane grafu, odpowiednio: ERDOS, CUBIC, GRID, BRIDGE, SMALL_WORLD, nazwa pliku, liczba wierzchołków i krawędzi. Wygenerowany graf zapisywany jest do pliku o podanej nazwie.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def gen_graph_to_file(graph_name,V,E,file_name):
    """"V,E-number of vertices and number of edges. For some graphs number of edges may differ""""
    G=None
    if graph_name=='ERDOS':
        G=nx.generators.random_graphs.erdos_renyi_graph(V,0.5)
    if graph_name=='CUBIC':
        G=nx.generators.small.cubical_graph()
    if graph_name=='GRID':
        G=nx.generators.lattice.grid_graph(dim=[V,V])
        V_dict={v:i for i,v in enumerate(list(G.nodes))}
        G=nx.relabel_nodes(G,V_dict)
    if graph_name=='BRIDGE':
        G2=nx.generators.random_graphs.erdos_renyi_graph(V//2,0.9)
        G1=nx.generators.random_graphs.erdos_renyi_graph(V//2,0.9)
        V_dict={v:i+(V//2) for i,v in enumerate(list(G1.nodes))}
        G1=nx.relabel_nodes(G1,V_dict)
        G=nx.compose(G2,G1)
        G.add_edge((V//2)-1,(V//2)+1)
    if graph_name=='SMALL_WORLD':
        G=nx.generators.geometric.navigable_small_world_graph(int(np.sqrt(V)))
        V_dict={v:i for i,v in enumerate(list(G.nodes))}
        G=nx.relabel_nodes(G,V_dict)
    file=open(file_name,"w")
    edges=list(nx.generate_edgelist(G,data=False))
    file.write(str(len(G.nodes))+' '+str(len(edges))+'\n')
    for line in edges:
        file.write(line+' '+str(round(np.random.uniform(0,5),3)))
        file.write('\n')
    file.write(edges[0]+' '+str(round(np.random.uniform(0,5),3)))
gen_graph_to_file('GRID',10,0,'test1.txt')
```

```python
class Circuit:
    """class to read from file and solve circut"""
    def __init__(self):
        self.N=0
        self.M=0
        self.E=0
        self.G=nx.Graph()
        self.Resistance={}
        self.SEM={}
    def read_from_file(self,file_name):
        f=open(file_name,"r")
        l=list(map(int,f.readline().split()))
        N,R=l[0],l[1]
        self.N=N
        self.E=R
        self.M=1
        for i in range(R):
            line=list(f.readline().split())
            u,v,r=int(line[0]),int(line[1]),float(line[2])
            self.Resistance[(u,v)]=r
            self.G.add_edge(u,v)
        line=f.readline().split()
        u,v,r=int(line[0]),int(line[1]),float(line[2])
        self.SEM[(u,v)]=r
    def solve_KL(self):
        """Solve using kirchoff's laws.  Return dictionary with current on each edge"""
        cycles=nx.algorithms.cycles.cycle_basis(self.G)
        A=np.zeros((len(cycles)+self.N,self.E))
        Z=np.zeros((len(cycles)+self.N,1))
        edges_enum={}
        for i,e in enumerate(list(self.Resistance.keys())):
            edges_enum[e]=i
        edges_direction={}
        for row,cycle in enumerate(cycles):
            cycle.append(cycle[0])
            for i in range(len(cycle)-1):
                u,v=cycle[i],cycle[i+1]
                edges_direction[(u,v)]=(u,v)
                edges_direction[(v,u)]=(u,v)
        for e in self.Resistance:
            if e not in edges_direction:
                edges_direction[e]=e
                edges_direction[(e[1],e[0])]=(e[1],e[0])
        for row,cycle in enumerate(cycles):
            for i in range(len(cycle)-1):
                u,v=cycle[i],cycle[i+1]
                e=(u,v) if (u,v) in self.Resistance else (v,u)
                if (u,v)==edges_direction[(u,v)]:
                    A[row][edges_enum[e]]=self.Resistance[e]
                elif (v,u)==edges_direction[(u,v)]:
                    A[row][edges_enum[e]]=-self.Resistance[e]
                if (u,v) in self.SEM:
                    Z[row]=self.SEM[(u,v)]
                if (v,u) in self.SEM:
                    Z[row]=self.SEM[(v,u)]
        for u in range(self.N):
            for v in self.G[u]:
                e=(u,v) if (u,v) in self.Resistance else (v,u)
                if (u,v)==edges_direction[(u,v)]:
                    A[u+len(cycles)][edges_enum[e]]=1
                elif (v,u)==edges_direction[(u,v)]:
                    A[u+len(cycles)][edges_enum[e]]=-1
        Z=np.linalg.lstsq(A,Z,rcond=None)[0]       
        return {edges_direction[e]:Z[edges_enum[e]][0] for e in edges_enum}
    def solve_MNA(self):
        """solve using node analysis. Return dictionary with current on each edge"""
        sem_val=list(self.SEM.values())[0]
        SEM_edge=list(self.SEM.keys())[0]
        SEM_edge_rev=(SEM_edge[1],SEM_edge[0])
        Resistance={(u,v):self.Resistance[(u,v)] for (u,v) in list(self.Resistance.keys())}
        Resistance[(SEM_edge[0],self.N)]=0
        Resistance[(self.N,SEM_edge[1])]=Resistance[SEM_edge if SEM_edge in Resistance else SEM_edge_rev]
        SEM={(SEM_edge[0],self.N):sem_val}
        Resistance.pop(SEM_edge if SEM_edge in Resistance else SEM_edge_rev)
        components = [(u,v,Resistance[(u,v)]) for u,v in Resistance]
        u_s=list(SEM.keys())[0][0]
        v_s=list(SEM.keys())[0][1]
        w_s=SEM[(u_s,v_s)]
        components.append((u_s,v_s,w_s))
        g2Count = 1
        matrixSize = self.N + 1
        A = np.zeros((matrixSize, matrixSize))
        b = np.zeros(matrixSize)
        g2Index = matrixSize - g2Count
        for i,component in enumerate(components):
            high = component[0]
            low = component[1]
            value = component[2]
            if i!=len(components)-1:
                if high != 0:
                    A[high - 1][high - 1] = A[high - 1][high - 1] + (1/value if value!=0 else 0)
                if low != 0:
                    A[low - 1][low - 1] = A[low - 1][low - 1] + (1/value if value!=0 else 0)
                if high != 0 and low != 0:
                    A[high - 1][low - 1] = A[high - 1][low - 1] - (1/value if value!=0 else 0)
                    A[low - 1][high - 1] = A[low - 1][high - 1] - (1/value if value!=0 else 0)
            else:
                if high != 0:
                    A[high - 1][g2Index] = A[high - 1][g2Index] + 1
                    A[g2Index][high - 1] = A[g2Index][high - 1] + 1
                if low != 0:
                    A[low - 1][g2Index] = A[low - 1][g2Index] - 1
                    A[g2Index][low - 1] = A[g2Index][low - 1] - 1
                b[g2Index] = value
                g2Index = g2Index + 1
#         print(A)
        sol={}
        x=np.array(0)
        x=np.append(x,np.linalg.solve(A, b))
        for (u,v) in Resistance:
            r=Resistance[(u,v)]
            if r!=0 and u!=self.N and v!=self.N:
                if x[u]>x[v]:
                    sol[(u,v)]=(x[u]-x[v])/r
                else:
                    sol[(v,u)]=(x[v]-x[u])/r
        SEM_edge=SEM_edge if x[self.N]>x[SEM_edge[1]] else (SEM_edge[1],SEM_edge[0])
        sol[SEM_edge]=x[self.N]-x[SEM_edge[0]] if x[self.N]>x[SEM_edge[0]] else x[SEM_edge[0]]-x[self.N] 
        return sol
    def plot(self,solve):
        """plot graph with current direction and value"""
        G1=nx.DiGraph()
        for e in solve:
            u,v=e[0],e[1]
            G1.add_edge(u,v)
            G1[u][v]['weight']=solve[(u,v)]
        poss=nx.circular_layout(G1)
        labels={}
        for e in G1.edges():
            labels[e]=round(G1[e[0]][e[1]]['weight'],3)
        s=0.0
        for v in G1.nodes:
            for u in G1[v]:
                s+=G1[v][u]['weight']
            for x in G1.nodes:
                if G1[x].get(v):
                    s-=G1[x].get(v).get('weight',0)
        if s<1*10**-12:
            print('OK')
        else:
            print('ERROR')
        plt.figure(3,figsize=(16,16)) 
        nx.draw_networkx_nodes(G1, poss)
        nx.draw_networkx_labels(G1, poss)
        nx.draw_networkx_edges(G1,poss,width=6,arrowsize=15,edge_cmap=plt.cm.Blues.reversed())
        nx.draw_networkx_edge_labels(G1,poss,labels)
```

```python
C=Circuit()
C.read_from_file('test1.txt')
solve_MNA=C.solve_MNA()
# solve_KL=C.solve_KL()
C.plot(solve_MNA)
```
